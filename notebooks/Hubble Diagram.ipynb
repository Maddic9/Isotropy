```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import gzip, cPickle
snFits = cPickle.load(gzip.GzipFile('snFits.p.gz'))

print len(snFits)
snf = [s for s in snFits.values() if s is not None]
print len(snf)
z = np.array([s.z for s in snf])
mu = np.array([s.mu for s in snf])

z = z[(mu > 0) & (mu < 19)]
mu = mu[(mu > 0) & (mu < 19)]
plt.scatter(z, mu)
plt.xlabel('z')
plt.xscale('log')
plt.ylabel('mu + const')
plt.xlim((0.2,1.4));
```
